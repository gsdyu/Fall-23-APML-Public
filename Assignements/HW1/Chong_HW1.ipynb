## Part 4

We are going to use `identity_matrix` to scramble the rows and columns of `img_average_channel`. 

1. First, create a numpy array that contains the numbers $0,1, \cdots, N$ where $N$ is the number of rows in the image.

2. Use `numpy.random.shuffle` to shuffle the indices. Call the shuffled array, `rows`

```python
some_row_matrix = np.array([[x] for x in range(image.shape[0])])
np.random.shuffle(some_row_matrix)
row = some_row_matrix
print(row)
```

## Part 5

1. Use your array of random row indices to create a new matrix `T` from `identity_matrix`. Overall, `T` has the same elements as `identity_matrix`, but the order of the rows of `T` should be according to the variable `rows`
2. Take a moment to think about what we are doing here. Describe in your own words what matrix `T` is.
3. Sanity check, print the shape of `T` and make sure it is the same size as `identity_matrix` and `img_average_channel`. If you need to,  reshape it to have the same size.


```python
T = identity_matrix.copy()
for i in range(image.shape[0]):
    T[i] = identity_matrix[row[i]]
'''2. What I know about matrix T is that it is a an identity_matrix but its rows are randomized so that its ones are no
      diagonal. At this moment it does not seem entirely clear what specific example this is, but judging from what I am
      given, this reminds me of real life data points where they themselves are randomized. The T matrix may be related to 
      a scenario like that where I am working with random and unsure data. I would also guess that T may be short for 
      transpose, which would make sense here as the rows are swapped around with one another. It may not however be
      directly related to a transpose matrix where its rows and column indices are swapped. Instead this is just some random
      swapping of the rows. T can also stand for transformed, which would also make sense as T is still the identity matrix
      but just slightly changed from the row swapping.'''
#Same shape for all 3 (T, identity_matrix, img_average_channel)
print(T.shape)
print(T.shape==identity_matrix.shape==img_average_channel.shape)
```

## Part 6

1. Matrix multiply `T` and `img_average_channel` and store the result in `transformed_image`. Check the shape of the result.
2. Plot `transformed_image` and describe the pattern you see. 
3. Think about what each dot product in the matrix multiplication is doing. Describe in your own words what this transformation is doing to the original image. 

```python
transformed_image = T@img_average_channel
plt.imshow(transformed_image)
'''2. The colors and level of brightness of the images looks similar/reused from the img_average_channel. However the pattern
      itself seem to be that there is no pattern. It looks completely randomized due to how uncomprehensible the image is.
      However, this type of image seem similar to static noise of an image.'''
'''3. Since a matrix times an identity matrix is just the matrix itself, and T is just the identity matrix but just the 
      rows swapped around, I would say that this transformation just swapped the row positions of the pixels. The T matrix, 
      being an identity, would not alter the values themselves.'''
```

## Part 7

1. We now want to retrieve the original image. Is matrix `T` invertible? How would you show that it is invertible?
2. Compute the inverse and store the matrix in `Tinv`.
3. Perform the inverse of the scrambling operation and plot the result to show that you retrieved the original image.

```python
'''1. A way to check if matrix T is invertible is to check the determinant of matrix T. If the determinant is not 0, then it is 
      invertible'''
print(np.linalg.det(T))
#T det is 1. Hence invertible.
Tinv = np.linalg.inv(T)
plt.imshow(Tinv@transformed_image)
```
